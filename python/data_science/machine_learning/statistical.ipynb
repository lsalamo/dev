# STATISTICAL

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Medidas de centralización
### Media aritmética

Se encuentra al sumar todos los datos y dividir entre el número de datos.

```python
data = np.array([10, 20, 30, 40, 50])
mean = np.mean(data)
print(f"Mean: {mean}")

mean = np.sum(data) / len(data)
print(f"Mean (Manual): {mean}")
```

### Mediana

Es el número de en medio; se encuentra al ordenar todos los puntos de datos y elegir el que está en medio (o si hay dos números en medio, tomar la media de esos dos números).

```python
# numero de elementos impares
data = np.array([10, 40, 20, 50, 30])
median = np.median(data)
print(f"Median: {median}")

# numero de elementos pares
data = np.array([10, 40, 20, 35, 50, 30])
median = np.median(data)
print(f"Median: {median}")

# manual
data = sorted(data)
n = len(data)
if n % 2 == 1:  # num elementos impares
    median = data[n // 2]
else:   # num elementos pares
    middle1 = data[n // 2 - 1]
    middle2 = data[n // 2]
    median = (middle1 + middle2) / 2
print(f"Median (Manually): {median}")
```

### Moda

Es el número más frecuente.

```python
data = np.array([1, 2, 2, 3, 2, 3, 4])

values, counts = np.unique(data, return_counts=True)
print(f"Frequency: {values}{counts}")
mode_index = np.argmax(counts)  # Index of the mode
mode = values[mode_index]  # Mode value
print("Mode:", values[mode_index])
print(f"Frequency: {counts[mode_index]}")

mode = stats.mode(data)
print(f"Mode: {mode.mode}")
print(f"Frequency: {mode.count}")
```

### Cuartiles, deciles y percentiles

CUARTILES
- Dividen la muestra ordenada en 4 partes iguales y hay 3 cuartiles.
- El primer cuartil C1 será la medida tal que el 25 % de los datos sean inferiores a su valor y el 75 % de los datos sean superiores. 
- El segundo cuartil C2 coincide con la mediana
- El tercer cuartil C3 marcará el valor tal que las tres cuartas partes de las observaciones sean inferiores a el y una cuarta parte sea superior.

DECILES
- Dividen la muestra ordenada en 10 partes iguales y 9 deciles

PERCENTILES
- Dividen la muestra en 100 partes iguales y 99 percentiles



```python
# Sample data (array)
data = np.array([10, 20, 30, 40, 50])

# Calculate the 25th, 50th, and 75th percentiles
percentiles = np.percentile(data, [25, 50, 75])

# Print the results
print(f"MIN: {np.min(data)}")
print(f"25th Percentile: {percentiles[0]}")
print(f"50th Percentile (Median): {percentiles[1]}")
print(f"50th Percentile (Median): {np.median(data)}")
print(f"75th Percentile: {percentiles[2]}")
print(f"MAX: {np.max(data)}")

# 90th Percentile
p90 = np.percentile(data, 90)
print("90th Percentile:", p90)

df = pd.DataFrame({'data': data})
print(df.describe())
```

## Medidas de dispersión 
### Desviación media

Es la media aritmética de los valores absolutos de lo que se desvía cada valor respecto a la media aritmética. Si las desviaciones son grandes la media no será tan representativa como cuando la desviaciones sean pequeñas.

$$\sigma_{media} = \frac{\sum_{i=1}^{n} |x_{i} - \bar{x}|}{n} = \frac{|x_{1} - \bar{x}| + |x_{2} - \bar{x}| + \cdots + |x_{n} - \bar{x}|}{n}$$

La siguiente tabla contiene las notas de 3 estudiantes en un curso con las asignaturas A, B y C. —> Todos los estudiantes tienen la misma media, pero, en qué caso la media representa mejor el rendimiento en el curso?



```python
# Example data
data = np.array([10, 20, 30, 40, 50])

# Calculate the mean
mean = np.mean(data)
print(f"Mean: {mean}")

# Subtract the mean from each data point
abs_diffs = np.abs(data - mean)
print(f"Absolute Differences:: {abs_diffs}")

# Calculate the median deviation
md = np.mean(abs_diffs)
print(f"Median Deviation: {md}")

md = np.sum(np.abs(data - mean)) / len(data)
print(f"Mean Deviation: {md}")
```

### Desviación típica o estandar
- La desviación estándar es una medida fundamental en estadística descriptiva y se utiliza para calcular la variación o distancia numérica de los datos con respecto a la media de un conjunto de datos.
- Usa los cuadrados para calcular las desviaciones, esto hace que los datos con desviaciones grandes influyan mucho en el resultado final.
- Una baja desviación indica que los datos están cerca de la media, mientras que una alta desviación sugiere que los datos están más dispersos

$$\sigma_{std} = \sqrt{\frac{\sum_{i=1}^{n} (x_{i} - \bar{x})^{2}}{n}} = \frac{(x_{1} - \bar{x})^{2} + (x_{2} - \bar{x})^{2} + \cdots + (x_{n} - \bar{x})^{2}}{n}$$




```python
# Example data
data = np.array([10, 20, 30, 40, 50])

# Calculate the standard deviation
std = np.std(data)
print(f"Standard Deviation (σ): {std}")

# Manual
mean = np.mean(data)
print(f"Mean: {mean}")

squared_diffs = (data - mean) ** 2
print(f"Squared Differences: {squared_diffs}")

variance = np.mean(squared_diffs)
print(f"Variance (σ^2): {variance}")

md = np.sqrt(variance)
print(f"Standard Deviation (σ) (Manual): {md}")
```

```python
# Plot data points
plt.figure(figsize=(6, 5))
plt.plot(data, 'o-', label='Data', color='blue')

# Plot the mean as a horizontal line
plt.axhline(y=mean, color='red', linestyle='--', label=f'Mean (μ = {mean:.2f})')

# Add lines for mean ± standard deviation
plt.axhline(mean + std, color='green', linestyle='--', label=f'Mean + Std Dev (μ + σ): {mean + std:.2f}')
plt.axhline(mean - std, color='green', linestyle='--', label=f'Mean - Std Dev (μ - σ): {mean - std:.2f}')

# Draw vertical lines from each data point to the mean
for i in range(len(data)):
    plt.vlines(i, ymin=min(data[i], mean), ymax=max(data[i], mean), color='blue', linestyle='dotted')

# Labeling
plt.title("Data Points with Mean, Standard Deviation and Variance Visualization")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
# Generate values for the x-axis (for the Gaussian curve)
mean = 30
std = 8
#  np.linspace generates 1000 points between  μ - 4σ  and  μ + 4σ 
# x = np.linspace(mean - 4*std, mean + 2*std, 40)
x = np.random.normal(mean, std, 25)
x = sorted(x)
mean = np.mean(x)
std = np.std(x)

# Probability Density Function (PDF)- Calculate the Gaussian (Normal) curve
# Densidad: La altura de la curva en cualquier punto indica la densidad de concentración de los valores en torno a ese punto. Una mayor densidad significa que es más probable que los valores estén cerca de ese punto.
# La fórmula de la PDF gaussiana calcula la densidad (o probabilidad) de una variable aleatoria x para una media y una desviación típica dadas.
# El área bajo toda la curva es igual a 1, lo que significa que la probabilidad total en todos los valores de x suma 1.
y = stats.norm.pdf(x, mean, std)
print(np.sum(y))

# Plot the Gaussian curve
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='purple', label='Data points')
plt.plot(x, y, color='purple', label='Gaussian Curve')      

# Plot the mean as a vertical line
plt.axvline(x=mean, color='red', linestyle='--', label=f'Mean (μ = {mean:.2f})')
plt.axvline(mean + std, color='green', linestyle='--', label=f'μ + σ (σ = {std:.2f}): {mean + std:.2f}')
plt.axvline(mean - std, color='green', linestyle='--', label=f'μ - σ (σ = {std:.2f}): {mean - std:.2f}')
plt.axvline(mean + 2*std, color='purple', linestyle='--', label=f'μ + 2σ: {mean + 2*std:.2f}')
plt.axvline(mean - 2*std, color='purple', linestyle='--', label=f'μ - 2σ: {mean - 2*std:.2f}')
plt.axvline(mean + 3*std, color='blue', linestyle='--', label=f'μ + 3σ: {mean + 3*std:.2f}')
plt.axvline(mean - 3*std, color='blue', linestyle='--', label=f'μ - 3σ: {mean - 3*std:.2f}')

# Calculate the percentages within 1, 2, and 3 standard deviations
within_1_std_dev = np.sum((x >= mean - std) & (x <= mean + std)) / len(x) * 100

# 2 standard deviations (95%)
within_2_std_dev = np.sum((x >= mean - 2*std) & (x <= mean + 2*std)) / len(x) * 100

# 3 standard deviations (99.7%)
within_3_std_dev = np.sum((x >= mean - 3*std) & (x <= mean + 3*std)) / len(x) * 100

# Calculate the probability for the range [20, 40]
prob_range = stats.norm.cdf(40, mean, std) - stats.norm.cdf(20, mean, std)
print(f"Probability of x being in the range [20, 40]: {prob_range:.4f}")

print(f"Percentage within (σ): {within_1_std_dev:.2f}%")
print(f"Percentage within (2σ): {within_2_std_dev:.2f}%")
print(f"Percentage within (3σ): {within_3_std_dev:.2f}%")

plt.title('Data Distribution with Gaussian Curve, Mean, and Standard Deviation')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

### Rango
El **rango**, también llamado **span**, es la distancia entre el mínimo y el máximo de una distribución, es decir, la distancia entre el valor más pequeño y el más grande. 

$$R = x_{max} - x_{min}$$



```python
# Example data
data = np.array([10, 20, 30, 40, 50])

# Calculate the standard deviation
range = np.max(data) - np.min(data)
print(f"Range (R): {range}")
```

### Interquartile Range (IQR)

Espectro en el que se situa el 50% medio de los valores, que corresponde a la diferencia entre el tercer cuartil (percentil 75) y el primer cuartil (percentil 25).



```python
# Example data
data = np.array([10, 20, 30, 40, 50])

# Calculate the interquartile range (IQR)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
print("Interquartile Range (IQR):", IQR)
```

```python
# Plot the data and metrics
plt.figure(figsize=(10, 6))

# Plot data points
plt.plot(data, 'o-', label='Data', color='blue')

# Add a horizontal line for the mean
mean= np.mean(data)
plt.axhline(mean, color='red', linestyle='--', label=f'Mean (μ): {mean:.2f}')

# Add lines for Range (max and min)
plt.axhline(np.max(data), color='purple', linestyle='--', label=f'Max: {np.max(data)}')
plt.axhline(np.min(data), color='purple', linestyle='--', label=f'Min: {np.min(data)}')

# Add lines for IQR (Q1 and Q3)
plt.axhline(Q1, color='green', linestyle='--', label=f'Q1 (25th percentile): {Q1:.2f}')
plt.axhline(Q3, color='green', linestyle='--', label=f'Q3 (75th percentile): {Q3:.2f}')

plt.title('Data with Dispersion Metrics')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

plt.show()
```
